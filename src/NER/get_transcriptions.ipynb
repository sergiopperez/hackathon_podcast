Especifica el link de la playlist:

```python
from youtubesearchpython import Search, Playlist


playlistVideos = Playlist.getVideos('https://www.youtube.com/playlist?list=PLLbN7SMQhMVZoXhtQ00AyebQE_-ttDrs9')
urls = []

for v in range(len(playlistVideos['videos'])):
    urls.append(playlistVideos['videos'][v]['link'])


print(f"Los vídeos contenidos en esta playlist son {urls}")
```


Reformatea las urls para que no incluyan información de la playlist:

```python
new_urls = []

for item in urls:
    idx = item.find("&list")
    if idx != -1:
        new_urls.append(item[:idx])
    else:
        new_urls.append(item)

print(new_urls)
```

Descarga los archivos mp3 y colocalos en una carpeta determinada:

```python
import yt_dlp

# specify output folder

audio_folder = '/content/drive/MyDrive/hackathon_podcast/data/NER/audio/'

for i in range(12): # Download only the first 12 audios 

  print(f'Downloading audio {i} out of 12')

  output_path = audio_folder + "output_" + str(i)
  ydl_opts = {
      'format': 'm4a/bestaudio/best',
      'outtmpl': output_path,
      # ℹ️ See help(yt_dlp.postprocessor) for a list of available Postprocessors and their arguments
      'postprocessors': [{  # Extract audio using ffmpeg
          'key': 'FFmpegExtractAudio',
          'preferredcodec': 'mp3',
      }]
  }

  with yt_dlp.YoutubeDL(ydl_opts) as ydl:
      ydl.download(new_urls[i])
```

Reformatea los archivos txt para dividirlos en frases de máximo 512 caracteres y mínimo 30:

```python
processed_transcript_folder = "/content/drive/MyDrive/hackathon_podcast/data/NER/processed_transcription/"

import nltk

# Download the necessary NLTK data if it's not already available
nltk.download('punkt')

for i in range(12):

  print(f'Processing raw transcript {i} out of 12')

  # Open the input file and read the text
  with open(f'{raw_transcript_folder}output_{i}.txt', 'r') as f:
      text = f.read()

  # Tokenize the text into individual sentences
  sentences = nltk.sent_tokenize(text)

  # Set the minimum and maximum number of characters per line
  min_chars_per_line = 30
  max_chars_per_line = 512

  # Split each sentence into lines based on the character limits
  output_lines = []
  for sentence in sentences:
      words = sentence.split()
      lines = []
      current_line = ''
      for word in words:
          if len(current_line) + len(word) + 1 > max_chars_per_line:
              # If adding the next word would exceed the maximum
              # line length, start a new line
              lines.append(current_line)
              current_line = word
          elif current_line == '':
              # If we're starting a new line, add the current word
              # without a space
              current_line = word
          else:
              # Otherwise, add the current word with a space
              current_line += ' ' + word
      if current_line:
          # Add the last line to the list of lines
          lines.append(current_line)
      # Add the lines for this sentence to the output
      if len(lines[0]) > min_chars_per_line:
        output_lines.extend(lines)

  # Join the output lines with newline characters as separators
  output_text = '\n'.join(output_lines)

  # Open the output file and write the text
  with open(f'{processed_transcript_folder}output_{i}.txt', 'w') as f:
      f.write(output_text)
```
